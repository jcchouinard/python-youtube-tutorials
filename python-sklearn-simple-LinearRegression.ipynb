## What is LinearRegression

`LinearRegression` is a ordinary least squares (OLS) Linear Regression from Scikit-Learn for predicting values. 

The `sklearn.linear_model.LinearRegression` class fits a linear model to minimize the residual sum of squares (RSS) between observations and predictions on a target.

##  Example on the Diabetes dataset

- Features (X): 10 physiological variables (age, sex, body mass index, blood pressure, etc.)
- Target (y): A quantitative measure of disease progression after one year.

```python
# Load data
from sklearn.datasets import load_diabetes
 
X, y = load_diabetes(return_X_y=True, as_frame=True)

print('Features:', X.columns)
```

```python
# Keep only the BMI predictor
X = X[['bmi']]
```

## Split the dataset into training and test sets

```python
# Split the dataset into training and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
                                    X, y, 
                                    test_size=0.3, 
                                    random_state=42)
```

## Training the LinearRegression

```python
# import LinearRegression
from sklearn.linear_model import LinearRegression

# train the model
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
y_pred = lin_reg.predict(X_test)
```

## Show the Linear Regression Intercept and Coefficients

```python
# Show Intercept and Coefficients
intercept = lin_reg.intercept_
coefficient = lin_reg.coef_[0]
print('Intercept:', intercept)
print('Coefficient for BMI:', coefficient) # Disease progression for 1 unit of BMI
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Plot the Linear Regression line with the data points
sns.regplot(x=X_test['bmi'], y=y_test)
plt.xlabel('Body Mass Index (BMI)')
plt.ylabel('Disease Progression')
plt.title('Linear Regression of BMI vs. Disease Progression')
plt.show()

```

```python
from sklearn import metrics
import numpy as np

print('Mean absolute error:', metrics.mean_absolute_error(y_test, y_pred))
print(f'Range of disease progession: {np.min(y)}, {np.max(y)}')
print('Root Mean squared error:', metrics.root_mean_squared_error(y_test, y_pred))
print('Coefficient of determination (r2 Score):', metrics.r2_score(y_test, y_pred))
```

## Interpretation
- Mean Absolute Error (MAE: 50.59): Average absolute difference between predicted and actual values. 50.59 units, moderate but notable error.
- Root Mean Squared Error (RMSE: 62.33): Average magnitude of the error. 62.33 units is somewhat accurate but the error is quite high. 
- Coefficient of Determination (R: 0.28): Only 28% of the variance in disease progression is explained by BMI. This is a a weak relationship. Need additional feature for better performance
